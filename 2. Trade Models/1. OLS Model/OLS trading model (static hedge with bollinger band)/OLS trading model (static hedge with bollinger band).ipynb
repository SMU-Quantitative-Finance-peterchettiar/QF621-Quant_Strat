```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
import math as math
%matplotlib inline
from sklearn.linear_model import LinearRegression
```

```python
final_pairs = pd.read_excel('final_pairs.xlsx')
```

```python
final_pairs
```

```python
rawdata = pd.read_excel('commod_data.xlsx')
```

```python
rawdata = rawdata.set_index('Date')
```

```python
rawdata
```

```python
traindata = rawdata.loc[:'2017-12-31',]
```

```python
testdata = rawdata.loc['2018-01-01':,]
```

```python
testdata
```

```python
return_df = (testdata / testdata.shift(1)) - 1
```

```python
for i in range(len(final_pairs)):
    X1 = final_pairs['Leg1'][i]
    X2 = final_pairs['Leg2'][i]

    x = traindata[X1].array.reshape((-1, 1))
    y = traindata[X2].array

    reg = LinearRegression().fit(x,y)

    beta = reg.coef_
    alpha = reg.intercept_

    Spread = rawdata[X1] - beta*rawdata[X2] - alpha

    df = pd.DataFrame()
    df[X1] = rawdata[X1]
    df[X2] = rawdata[X2]
    df['Spread'] = Spread

    #set window to 20
    window= 20

    df['Spread SMA'] = df['Spread'].rolling(window = window).mean()
    df['Spread Vol'] = df['Spread'].rolling(window = window).std()
    
    #setting 2 SD spread vol for upper and lower bb
    df['upper_bb'] = df['Spread SMA'] + df['Spread Vol'] * 2
    df['lower_bb'] = df['Spread SMA'] - df['Spread Vol'] * 2

    df = df.loc["2018-01-01":,]

    #simple mean-reversion strategy without stop loss
#     df['position'] = 0
#     df.loc[df['Spread']> df['upper_bb'], 'position'] = -1
#     df.loc[df['Spread']< -df['lower_bb'], 'position'] = 1

    #Using woon's stop loss strategy
    df['position'] = 0
    df.loc[df['Spread']> 1*df['upper_bb'], 'position'] = -2.5
    df.loc[df['Spread']< 1*-df['lower_bb'], 'position'] = 2.5
    df.loc[df['Spread']> 2*df['upper_bb'], 'position'] = -5
    df.loc[df['Spread']< 2*-df['lower_bb'], 'position'] = 5
    df.loc[df['Spread']> 3*df['upper_bb'], 'position'] = 0
    df.loc[df['Spread']< 3*-df['lower_bb'], 'position'] = 0
    
    df['X1X2position'] = df['position'].shift(1)
    df['X1position'] = df['X1X2position']
    df['X2position'] = df['X1X2position']*beta*-1
    df['dailypnl'] = df['X2position']*return_df[X2].shift(-1) + df['X1position']*return_df[X1].shift(-1)
    df['cumpnl'] = df['dailypnl'].cumsum()
    df['cumpnl'].plot(figsize=(10,6),label = X1 + X2)
    plt.xlabel("Date")
    plt.ylabel("PnL")
    plt.legend()
    plt.show()
```
