```python
import numpy as np
import statsmodels.api as sm
import statsmodels.tsa.stattools as ts
import pandas as pd

from sklearn.linear_model import LinearRegression
```

```python
def get_hurst_exponent(time_series, max_lag=20):
    """Returns the Hurst Exponent of the time series"""
    
    lags = range(2, max_lag)

    # variances of the lagged differences
    tau = [np.std(np.subtract(time_series[lag:], time_series[:-lag])) for lag in lags]

    # calculate the slope of the log plot -> the Hurst Exponent
    reg = np.polyfit(np.log(lags), np.log(tau), 1)

    return reg[0]
```

```python
def apply_half_life(time_series):
    lag = np.roll(time_series, 1)
    lag[0] = 0
    ret = time_series - lag
    ret[0] = 0
    
    # adds intercept terms to X variable for regression
    lag2 = sm.add_constant(lag)

    model = sm.OLS(ret, lag2)
    res = model.fit()

    half_life = -np.log(2) / res.params[1]    
    
    return half_life
```

```python
data = pd.read_excel("commod_data.xlsx")
```

```python
data = data.loc[(data["Date"] >= '2009-01-01')
                     & (data["Date"] < '2017-12-31')]
```

```python
pairs = pd.read_excel("commod_list.xlsx")
```

```python
Leg1 = pairs["Leg 1"].tolist()
Leg2 = pairs["Leg 2"].tolist()
```

```python
df = []

for n in Leg1:
    for m in Leg2:
        x = data["%s" % n].array.reshape((-1, 1))
        y = data["%s" % m].array
             
        reg = LinearRegression().fit(x,y)
        Alpha = reg.intercept_
        Hedge = 1/reg.coef_
        
        
        Spread = data["%s" % m]*Hedge - data["%s" % n] - Alpha
        Spread = np.array(Spread).reshape((-1,1))
        Spread = pd.DataFrame(Spread)
        
        hurst_exp = get_hurst_exponent(Spread, 64)
        
        half_life = apply_half_life(Spread[0])
        
        df.append([["%s" % m], ["%s" % n], Hedge, hurst_exp, half_life])
        
df = pd.DataFrame(df, columns=['Leg1', 
                               'Leg2',
                               'Hedge Ratio',
                               'Hurst Exponent',
                               'Half Life'])
```

```python
df.to_excel("hurst exponent & half life_pairs.xlsx")
```
